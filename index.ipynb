```python
data = pd.read_csv('.csv')
```

```python
# Classify the original table by ID and calculate the total intake of each food item corresponding to each ID
data_sum = data.groupby(['Respondent ID']).sum()
data_sum.head()
```

```python
# The classification in the table should be divided into three categories—healthy foods, unhealthy foods, and meat-based foods—to facilitate the subsequent calculation of the index.
healthy_plant = [
    'Whole grain', 'Fruits', 'Vegetable',
    'Nuts', 'Legumes', 'Vegetable Oil', 'tea_coffee'
]

unhealthy_plant = [
    'Juice fruit', 'Added sugars', 'Refined grain',
    'Potatoes', 'Alcohol'
]

animal_foods = [
    'Solid fat', 'Dairy', 'Eggs',
    'Seafood', 'Meat', 'misc_animal_foods'
]
```

```python
scores = pd.DataFrame(index=data_sum.index)

# hPDI:
for col in data_sum.columns:
    quintiles = pd.qcut(data_sum[col], q=5, labels=False, duplicates='drop') + 1  # Q1=1, Q5=5
    if col in healthy_plant:
        scores[col] = quintiles  # Positive scoring
    else:
        scores[col] = 6 - quintiles  # Negative scoring

# Calculate the total score of hPDI
data_sum['hPDI'] = scores.sum(axis=1)

# uPDI:
scores_u = pd.DataFrame(index=data_sum.index)
for col in data_sum.columns:
    quintiles = pd.qcut(data_sum[col], q=5, labels=False, duplicates='drop') + 1
    if col in unhealthy_plant:
        scores_u[col] = quintiles  # Positive scoring
    else:
        scores_u[col] = 6 - quintiles  # Negative scoring

# Calculate the total score of uPDI
data_sum['uPDI'] = scores_u.sum(axis=1)

# Show the results
print(data_sum.describe())
```

```python
# Download the file
data_sum.to_csv('output.csv', encoding = 'utf-8-sig')
files.download('output.csv')
```
